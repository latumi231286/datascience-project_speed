```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

rootFolder = "/home/giangtm/Work/Projects/DataScience/dataset/"
d= pd.read_csv(rootFolder + "Project_Speed_Data_3.csv")
d.sort_values(['project_type', 'project_name', 'step'], ascending=[False, True, True], inplace=True)
d.drop(['different'], axis=1, inplace=True)
d.rename(columns={'real_speed': 'speed'}, inplace=True)
d.head(n=20)
```

```python
d['speed'].describe()
```

```python
d_gb_step = d.groupby(['project_name','step'])['speed']
d_gb_step.describe()
```

```python
step = 'Capture'
d_gb_step = d[d['step']==step].groupby(['project_name','step'])['speed']
db_des = d_gb_step.describe()
db_des.sort_values(['mean'],ascending=[False], inplace=True)
db_des.head(n=10)
```

```python
def drawStatisticByStep(step, no_of_ele, color):
# group by step
    d_gb_step = d[d['step']==step].groupby(['project_name','step'])['speed']
# put values decribe
    db_des = d_gb_step.describe()
# sort
    db_des.sort_values(['mean'],ascending=[False], inplace=True)
    
    d_max_10 = db_des.nlargest(no_of_ele, 'mean')
    d_max_10.sort_values(['mean'],ascending=[False], inplace=True)
    # ax1 = d_max_10[['min','std','max']].plot(linestyle='-', marker='o')
    t = d_max_10[['mean', 'std']].plot(kind='bar', figsize=(18,10), yerr='std', color=color)
    t.set_ylim(100,450);
    t.set_title(" Projects Lowest Speed In " + str(step), fontsize=18, color='blue')
    t.set_xlabel("Project", fontsize=8)
    t.set_ylabel("Time (seconds)", fontsize=10)
  
```

```python
drawStatisticByStep('Capture', 20, 'brown')  
```

```python
drawStatisticByStep('Classify', 20, 'green')  
```

```python
drawStatisticByStep('Classify', 20, 'gray') 
```
